```python
#Read in Training data
trainingDF = pd.read_csv('train.csv')
#Setting index to start at 1
trainingDF.index +=1
print("Training Data Summary")
#printing summary features: Shape
display(trainingDF.shape)
#Stats
display(trainingDF.describe())
#First 5 rows
display(trainingDF.head())
#type summary
trainingDF.info()
#column names
print(trainingDF.columns)
```

```python
#Check for null data
print("Training Data")
trainingDF.isnull().sum()
```

```python
#drop the target column from the training data X
X = trainingDF.drop(['target'], axis=1)
#Y is only the target column
y = trainingDF['target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
X_train.shape, X_test.shape
```

```python

cv=CountVectorizer(min_df=1,max_df=1,ngram_range=(1,2))
#transformed train reviews
cv_train=cv.fit_transform(X_train['text'])
#transformed test reviews
cv_test=cv.transform(X_test['text'])

print('cv_train:',cv_train.shape)
print('cv_test:',cv_test.shape)
```

```python
lr=LogisticRegression(penalty='l2',max_iter=500,C=1,random_state=0)
#Fitting the model for words
lr=lr.fit(cv_train,y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report, f1_score
# Make predictions
lr_predict = lr.predict(cv_test)

# This model caluclates F1 score for both classifications. Compute weighted F1 score
lr_f1_score = f1_score(y_test, lr_predict, average='weighted')
print("F1 Score:", lr_f1_score)
```
